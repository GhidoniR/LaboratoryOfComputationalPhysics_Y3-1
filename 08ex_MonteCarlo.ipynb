```python
tau=3.052*60

NTl=1000
NPb=0

time=0
p=1-2**(-1/tau)
print(p)
listNP=[]
listTL=[]
tot=[]
while NPb<1000 :
    time=time+1
    for i in range(NTl):
        l=np.random.uniform(0,1)
        if l<p:
            NTl=NTl-1
            NPb+=1
    listNP.append(NPb)
    listTL.append(NTl)
    tot.append(NPb+NTl)

        
    
x=[i for i in range(0,time)]
print(time)
plt.plot(x, listNP, label = 'Thallium')
plt.plot(x, listTL, label = 'Lead')
plt.plot(x, tot, label = 'total')
plt.legend()
plt.axis([0, time, 0, 1020])
plt.ylabel('numbers of atoms')
plt.xlabel('time')
plt.title('Atoms');
```
